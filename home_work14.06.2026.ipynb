```python
import numpy as np 
import pandas as pd
```

```python
Worker = ['Alice', 'Bob', 'David'] 
Hours = [8, 6, 7]
Monday = pd.Series(data = Hours, index = Worker)
print(f'Monday \n{Monday}')

Tuesday_dict = {
    'Bob':    8,
    'Carol': 5,
    'David': 6
}
Tuesday = pd.Series(Tuesday_dict)
print(f'\nTuesday \n{Tuesday}')

Wednesday_dict = {
    'Alice':    7,
    'Carol': 8,
    'David': 5
}
Wednesday = pd.Series(Wednesday_dict)
print(f'\nWednesday \n{Wednesday}')
```

```python
total_hours = Monday.add(Tuesday , fill_value=0).add(Wednesday  , fill_value=0)
print(total_hours)
```

```python
print(total_hours.idxmax(),total_hours.max())
```
